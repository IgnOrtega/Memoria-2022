```python
def generate_name_df(densidad,n,tipo_grafo,parametro):
    if tipo_grafo=="escala":
        alpha="m"+str(int(parametro))        
    else:
        alpha="p"+str(int(1/parametro))    
    name_graph ="densidad_"+densidad+"graph_n"+str(n)+alpha+"n°"
    name_df="data frame/caso_sin_falsos/"+"s_7_8_"+name_graph
    return name_df
```
